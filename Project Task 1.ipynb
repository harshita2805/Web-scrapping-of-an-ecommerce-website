```python
"https://yoshops.com/t/toys?page=2"
Yoshops_link = "https://yoshops.com/t/toys?page=2"
uClient = uReq(Yoshops_link)
Yoshops_page = uClient.read()
Yoshops_html = bs(Yoshops_page, "html.parser")
bigboxes = Yoshops_html.findAll("div", {"class": "product"})
range_=len(bigboxes)
product_link=[]
for i in range(range_):
    big =bigboxes[i]
    product =big.a['href']
    full_link = Yoshops_link+product
    product_link.append(full_link)
product_link
pro_name=[]
range_=len(bigboxes)
for j in range(range_):
    big =bigboxes[j]
    product_name=big.find("span",attrs={"class": 'product-title'}).string
    pro_name.append(product_name)
pro_name
image=[]
range_=len(bigboxes)
for i in range(range_):
    big =bigboxes[i]
    image_link =big.div.div.img["src"]
    image.append(image_link)
#print(image)
#print(pro_name)
#print(product_link)
data = {"Product Name":pro_name,"Product Link":product_link, "Image Link":image}
result = pd.DataFrame([pro_name,product_link,image])
#save_file = result.to_csv("C:\Users\Windows\Desktop\Yoshops Intership\app.csv")
print(result)
```
